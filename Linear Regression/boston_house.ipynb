# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x = df.drop(['Price'],axis=1)
y = df['Price']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state = 4)
```

```python
df.head()
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
coef = pd.DataFrame([x_train.columns,lr.coef_]).T
coef = coef.rename(columns={0:'Attribute',1:'Coefficients'})
coef
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print('MSE : ',mean_squared_error(y_test,y_pred))
print('MAE : ',mean_absolute_error(y_test,y_pred))
print('RMSE : ',mean_squared_error(y_test,y_pred)**0.5)
```

```python
fig = plt.figure(figsize=(10,8))
plt.scatter(x=np.linspace(0,50,152),y=y_pred,c='r')
plt.scatter(x=np.linspace(0,50,152),y=y_test,c='b')
# plt.plot(np.linspace(0,50,152),y_test-y_pred,c='r')
sns.set_style(style='darkgrid')
plt.show()
```

```python
print('Error Percent : {:.2f} %'.format(np.abs(np.mean(y_test-y_pred)*100)))
```

# SVM REGRESSOR

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn import svm
reg = svm.SVR()
reg.fit(x_train,y_train)
```

```python
def print_error(predictions,actual):
    print('MSE : ',mean_squared_error(predictions,actual))
    print('MAE : ',mean_absolute_error(predictions,actual))
    print('RMSE : ',mean_squared_error(predictions,actual)*0.5)
    print(f'Error Percent : {np.abs(np.mean(actual-predictions)):.2f}%')
```

```python
y_svr_pred = reg.predict(x_test)
print_error(y_svr_pred,y_test)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor()
rf_reg.fit(x_train,y_train)
```

```python
y_rf_pred = rf_reg.predict(x_test)
print_error(y_rf_pred,y_test)
```
